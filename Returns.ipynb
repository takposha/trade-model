```python
import pandas as pd
import numpy as np
from statistics import *
save = []
best = [0,1, np.random.rand(10)]

IB = pd.read_excel("IB10.xlsx", 
                    sheet_name='Sheet1')
names = list(filter(None, IB.tail(1).values[0]))[1:][1::2]

IB = IB.drop('Date', 1)
IB = IB[:-1]
IB = IB.values

#IB
```

```python
returnsperSto = []
for j in range(0,np.shape(IB)[1],2):
    ret = []
    for i in range(0,np.shape(IB)[0]):
        ret.append((IB[i][j+1]-IB[i][j])/IB[i][j])
    returnsperSto.append(ret)
Data = pd.DataFrame(np.array(returnsperSto).T, columns=names)
#display(Data)

returnsPerAnnum = []
for j in range(len(returnsperSto)):
    ret = []
    for i in range(0,len(returnsperSto[0])-1,12):
        #print(i)
        retAnn = 1
        for k in range(i,i+11):
            #print(k)
            retAnn = retAnn*(1+returnsperSto[j][k])
        ret.append((retAnn-1)*100)
    returnsPerAnnum.append(ret)
    
AnnData = pd.DataFrame(np.array(returnsPerAnnum).T, columns=names)
display(AnnData)

length = len(returnsPerAnnum)
ret = []
returns = np.empty(length)
risk = np.empty(length)
variance = np.empty(length)
table = np.empty([length,length])
weights = np.empty(length)
someStep = np.empty(length)

'''
for i in range(length):
    ret.append(np.array(IB[names[i]]))
    returns[i] = np.mean(ret[i])
    risk[i] = np.std(ret[i])
    variance[i] = np.var(ret[i])
''' 
for i in range(length):
    ret.append(np.array(returnsPerAnnum[i]))
    returns[i] = np.mean(ret[i])
    risk[i] = np.std(ret[i])
    variance[i] = np.var(ret[i])

for i in range(length):
    for j in range(length):
        if i != j:
            table[i][j] = np.cov(ret[i],ret[j])[0][1]
            table[j][i] = table[i][j] 
        else:
            table[i][j] = np.var(ret[i])
            
display(pd.DataFrame(table))

display(pd.DataFrame(returns))
display(pd.DataFrame(risk))
display(pd.DataFrame(variance))
```

```python
e = np.ones(length)
C = np.asmatrix(table)
y = np.linalg.solve(C,e)
Y = y/np.sum(y)
weights = Y

weights = np.array([0.39, 0.1, 0.001 , 0.01, 0.03, \
         0.33, 0.02, 0.016, 0.001, 0.1])

weights = np.array([0.39, 0.1, 0.0 , 0.01, 0.03, \
         0.33, 0.02, 0.02, 0.0, 0.1])

print(np.sum(weights))
```

```python
for i in range(length):
    someStep[i] = np.sum(weights*table[i])
    
portVar =  np.sum(weights*someStep)
portWeights = np.sum(weights*returns)

print('Portfolio Returns: ',portWeights*100,'%')
print('Portfolio Risk: ',portVar**0.5*100,'%')
for i in range(length):
    print(names[i],':', weights[i])
```

```python
best = [0,50, np.zeros(10)]

for j in range(1000000):
    weights = np.random.rand(10)
    weights = weights / np.sum(weights)
    for i in range(length):
        someStep[i] = np.sum(weights*table[i])

    portWeights = np.sum(weights*returns)*100
    portVar =  np.sum(weights*someStep)**0.5*100
    
    if portWeights>best[0]:
        if portVar<best[1]:
                best = [portWeights,portVar, weights]
    
    #print('Portfolio Returns: ',portWeights,'%')
    #print('Portfolio Risk: ',portVar,'%')
    
print(best)
```
